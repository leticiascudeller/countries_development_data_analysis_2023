```python
import pandas as pd
countries = pd.read_csv('data.csv')
countries.head()
```

```python
#Let's firstly check if there's any missing values and what are the types of each column's value
countries.info()
```

```python
#Let's add the missing 'e' in AveraScore column name
column_mapper = {'AveragScore': 'AverageScore'}
countries = countries.rename(mapper = column_mapper ,
            axis=1)
```

```python
#Which countries rank the highest and lowest in overall?
countries.sort_values(by='AverageScore', ascending=False)
```

```python
#Which country is closest to the mean average score across all indicators?
countries['AverageScore'].describe()
#I will discover the country's name with the closest value to mean average. 
closest_avg_inferior = countries['AverageScore'] <= 58.06
countries[closest_avg_inferior]
```

```python
#I will discover the country's name with the closest value to mean average. 
closest_avg_above = countries['AverageScore'] >= 58.06
countries[closest_avg_above]
```

```python
#We can conclude that Colombia has the closest average score to the mean average between all countries.
```

```python
#Which countries have the highest and lowest SafetySecurity scores?
countries.sort_values(by ='SafetySecurity',ascending=False)
#Highest: Luxembourg \ Lowest: South Sudan
#It is possible to notice that the countries with the highest average score tend to have a Safety/Security score even higher while the opposite happens with countrie's with the lowest scores
```

```python
#Which regions or countries excel in PersonalFreedom, and which lag behind?
countries.sort_values(by='PersonelFreedom',ascending=False)

#Highest:Norway \ Lowest: Sytya
```

```python
#Is there a relationship between PersonalFreedom and Governance scores?
column_mapper = {'PersonelFreedom': 'PersonalFreedom'}
countries = countries.rename(mapper = column_mapper ,
            axis=1)
countries.sort_values(by='Governance',ascending=False)
#there is a clear link between a governance score and personal freedom
```
